```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

## Plotting the Rejection Region

```python
# Generate x values for the standard normal distribution
x=np.linspace(-4,4,1000)


#calculate teh critcal z-scores
z_critical=stats.norm.ppf(1-0.05/2)

#plot  the standard normal distribution
plt.figure(figsize=(8,6))
plt.plot(x,stats.norm.pdf(x), 'black', lw=2, label='Standard distribution')

#Shade the rejection region
plt.fill_between(x,stats.norm.pdf(x),where=(x<= -z_critical) | (x>=z_critical), color='red',alpha=0.5,label='Rejection Region')
                 
#add Labels and title                
plt.title('stnadard normal distribution with rejecton region)')
plt.xlabel('z-score')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## Perform test statistic and Calculate P-value

```python
# sample data weights
sample=[150, 155,160,165,170, 175, 180]
 
#sample statistics

sample_mean=np.mean(sample)
sample_std=np.std(sample, ddof=1)
sample_size=len(sample)

#hypothesized mean
mu_0=165

#calculate the z-score
z_score=(sample_mean-mu_0)/(sample_std/np.sqrt(sample_size))
    
#calculate the p-value
p_value=2 * (1-stats.norm.cdf(np.abs(z_score)))
```
