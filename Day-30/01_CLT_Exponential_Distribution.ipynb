```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exponential Distribution

```python
# Set the seed for reproductivity
np.random.seed(42)

#Generate the population distribution
population=np.random.exponential(scale=1,size=10000) #scale =gamma=1/beta; beta=rate of change

#plot the population distribution
plt.figure(figsize=(8,6))
plt.hist(population, bins=50,density=True, color='blue',alpha=0.5)
plt.title("Population Distribution(Exponential)")
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
