```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Generate random numbers from a skewed normal distribution
x = np.random.standard_t(10, size=1000)
x[:15]
```

```python
plt.hist(x, color = 'blue', alpha=0.5)
```

# Mean 

```python
mean = np.mean(x)
print("Mean:", mean)
```

```python
plt.hist(x, color = 'green',alpha=0.5)
plt.axvline(x = mean, color='black',alpha=1)
```

# Median 

```python
median=np.median(x)
print("Median",median,)
```

```python
plt.hist(x, color = '#33DFDA',alpha=0.9)
plt.axvline(x = median, color='black')
```

# Mode

```python
mode = stats.mode(x).mode
print("Mode:", mode)
#mode=stats.mode(x).mode
#print("Mode",mode)
```

```python
plt.hist(x, color = '#1980DA')
plt.axvline(x = mode, color='black')
```
