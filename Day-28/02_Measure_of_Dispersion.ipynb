```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random numbers from a skewed normal distribution
x=np.random.standard_t(10,size=1000)
x[:15]
```

```python
plt.hist(x, color='blue',alpha=0.5)
```

# Variance

```python
variance=np.var(x)
print("Variance: ",variance)
```

```python
plt.hist(x,color="#7C9B79")
plt.axvline(x=x.mean(),color='black')
plt.axvline(x=variance,color='yellow')
```

# Stsandard Deviation

```python
std=np.std(x)
print("Standard Deviation: ", std)
```

```python
plt.hist(x,color='blue',alpha=0.5)
plt.axvline(x=std,color='black')
plt.axvline(x=x.mean(),color='yellow')
```
